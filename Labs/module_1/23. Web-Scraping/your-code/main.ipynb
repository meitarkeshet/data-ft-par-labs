#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
# This is the url you will scrape in this exercise
page = requests.get(url) # download the website page using "get" into a var called 'page'
#page.status_code # 200 indicates sucsses 
soup = BeautifulSoup(page.content, 'html.parser') #

center = soup.select("[lang]")

#[i.get('strong') for i in center]
center = [i.text for i in center][1:]
center = [i for i in center if re.search("\n", i)]
lang = [ i.replace("\n","") for i in center ]
#lang = lang.match
lang
# until a string
````

````python
lang_ex = [re.split('[0-9]', str(i))[0] for i in lang]
lang_ex
````

````python
lang
lang_num = ["".join(re.findall(r'[0-9]', str(i))) for i in lang]
merged =[(lang_ex[i], lang_num[i]) for i in range(0, len(lang_num))] 
merged
````
