```python
# clean columns
df_dropped = df_final.drop(columns=['currenciesAccepted', 'hasLoyaltyProgram', 'isBookable',
                                   'isInsider','uuid', 'canBurnYums','highlightedTag',
                                   'marketingOffer.type','bestOffer.tagsIds','bestOffer.name',
                                   'yumsDetail.isSuperYums','yumsDetail.description',
                                   'yumsDetail.yumsCount','marketingOffer', 'snippets',
                                   'photosCarousel','id', 'geo.longitude', 'geo.latitude',
                                   'mainPhotoSrc','mainPhoto.664x374','mainPhoto.1350x759',
                                   'mainPhoto.160x120','mainPhoto.80x60','mainPhoto.92x92',
                                   'mainPhoto.184x184', 'bestOffer.id', 'mainPhoto.source',
                                   'mainPhoto.612x344', 'mainPhoto.480x270','address.country',
                                   'address.locality'])
df_dropped
```

```python
#df_dropped.loc[:, 'aggregateRatings.thefork.ratingValue']
ratings = df_dropped[['aggregateRatings.thefork.ratingValue', 'aggregateRatings.tripadvisor.ratingValue']] 
# format the fork rating (1-10) so that it works with the tripadvisor (1-5)
ratings['aggregateRatings.thefork.ratingValue'] /= 2
# avrage each row between the for rating and tripadvisor rating column
ratings = ratings.mean(axis=1)
#add the ratings we calculated as a column in the data_frame
df_dropped["ratings"] = ratings
df_dropped
```

```python
# drop the now not-relevent ratings
cln_dropped = df_dropped.drop(columns=['aggregateRatings.thefork.ratingValue',
                                      'aggregateRatings.thefork.reviewCount',
                                      'aggregateRatings.tripadvisor.ratingValue',
                                      'aggregateRatings.tripadvisor.reviewCount'])
# extract the quartier from the code postal
quartier = cln_dropped['address.postalCode']
quartier = [(int(i) % 100) for i in list(quartier)]
# add to data frame
cln_dropped["quartier"] = quartier
# drop the now unused postal code
cln_dropped = cln_dropped.drop(columns=["address.postalCode"])
cln_dropped
```

```python
cln_dropped
# calculate the "offer"'s' effect on the "priceRange"
# create lists 
pricerange = list(cln_dropped['priceRange'])
offer = list(cln_dropped['bestOffer.label'])
# turn empty values ("") and "MENU" into 0
# clean the rest
offer = list([
        i.replace("", "0") if i == ''
        else
        (i.replace("%", "").
         replace("-", "").
         replace("MENU", "0")
         )
         for i in offer])
# turn into ints
offer = [int(i) for i in offer]

#reduce offer from price range

afterReduc = []

#notice the rounding

for i in list(range(len(offer))):
    afterReduc.append(round((pricerange[i] - ((offer[i] / 100)*pricerange[i]))))

# add the column to a new dataFrame
    
byPrice = cln_dropped
byPrice["priceRAfterReduc"] = afterReduc
byPrice
```

```python
# drop the price columns 
byPrice = byPrice.drop(columns=["bestOffer.label", "priceRange"])

# reorder 
byPrice = byPrice.reindex(columns= ['name', 'servesCuisine', 'priceRAfterReduc',
                                    'ratings','quartier','address.street',
                                    'marketingOffer.title','mainPhoto.240x135'
                                    ])
```
