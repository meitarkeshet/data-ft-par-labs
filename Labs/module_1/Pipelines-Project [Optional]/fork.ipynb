```python
FIELDS = ['name', 'priceRange', 'servesCuisine']
def clean_data(data):
    table = pd.DataFrame()
    for i in range(len(data) - 1):
        df = pd.json_normalize(data[i + 1])
        df = df[FIELDS]
        table = table.append(df)
    return table

cleand = clean_data(data)
cleand.shape
# problematic : 'address' , 'aggregateRatings', bestOffer
```
