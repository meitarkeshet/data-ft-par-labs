```python
companey_signed_duration_df = pd.DataFrame(columns=['company_id','days_of_service'])
for name,group in grouped:
    #print (name)
    tmp_group = group.signature_date.sort_values()
    first_sign = datetime.strptime(tmp_group.iloc[0].split(' ')[0], "%Y-%m-%d")
    last_sign = datetime.strptime(tmp_group.iloc[-1].split(' ')[0], "%Y-%m-%d")
    total_contracted = last_sign - first_sign
    #print(total_contracted.days)
    #print("breakkkkkkkkkk")
    companey_signed_duration_df=companey_signed_duration_df.append({'company_id': name, 'days_of_service': total_contracted.days}, ignore_index=True)
```
